```python
# 画图一
import os
import numpy as np
import matplotlib.pyplot as plt

message_dict = {}

def count_thread(DIR):
    return len([name for name in os.listdir(DIR) if os.path.isfile(os.path.join(DIR, name))])


def count_message(file):
    for line in file:
        line=line.strip('\n')
        temp_list = line.split(',')
        if len(temp_list) <= 1:
            continue
        func_no = int(temp_list[0])
        size = 0
        if func_no >= 3 and func_no <= 16:
            size = int(temp_list[1]) / 2 * mpi_datatype[temp_list[2]][1]
        elif func_no >= 17 and func_no <= 26:
            size = int(temp_list[1]) * mpi_datatype[temp_list[2]][1]
        elif func_no >= 27 and func_no <= 34:
            size = int(temp_list[1]) * mpi_datatype[temp_list[2]][1] * (thread_num-1)
        if size in message_dict:
            message_dict[size] += 1
        else:
            message_dict[size] = 1
    print(message_dict)

def sum_message(my_dict):
    sum = 0
    for i in my_dict: 
        sum = sum + my_dict[i] 
    return sum

def draw_bar(res):
    bin = int(max(res.keys()) / 20 / 100) * 100

    temp_dict = {}
    for key in res.keys():
        temp_key = int(key / bin) * bin
        if temp_key in temp_dict:
            temp_dict[temp_key] += res[key]
        else:
            temp_dict[temp_key] = res[key]
    print(temp_dict)
    x = []
    y = []
    for i in range(0, max(res.keys()), bin):
        x.append(str(i) + "-" + str(i+bin))
        if i in temp_dict:
            y.append(temp_dict[i])
        else:
            y.append(0)
#     print(x)
#     print(y)
    plt.barh(x, y)
    plt.xlabel('message num')
    plt.ylabel('message size(bytes)')

    # plt.xticks(rotation=60) 
    plt.show()


    
    
    
path = "../data"
thread_num = count_thread(path)
files = os.listdir(path)
for file in files:
    filename = path + "/" + file
    with open(filename, 'r', encoding = 'utf-8') as f:
        count_message(f)

message_num = sum_message(message_dict)        
# print(message_num)
draw_bar(message_dict)

```
